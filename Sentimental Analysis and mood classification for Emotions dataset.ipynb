```python
import nltk # natural language tool kit
#nltk.download('stopwords') 
#used to import the stopwords
#it will downlload a text file which will have stop words
common_stopwords=nltk.corpus.stopwords.words('english')
```

```python
#5. LSTM (Long-Short Term Memory)
```

```python
#LSTM stands for Long-Short Term Memory. 
#LSTM is a type of recurrent neural network but is better than traditional recurrent neural networks in terms of memory. 
#Having a good hold over memorizing certain patterns LSTMs perform fairly better.
```

```python
Train3=Train1[['textID','selected_text','mood']]
Train3.head(2)#Extracting the important columns for Modelling
```

```python
Test3=Test1[['textID','selected_text','mood']]
Test3.head(2)#Extracting the important columns for Modelling
```

```python
Val3=Val1[['textID','selected_text','mood']]
Val3.head(2)#Extracting the important columns for Modelling
```

```python
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
#Assumptions
vocab_size = 1000
len_sentence=150
```

```python
#Train
```

```python
one_hot_word1 = [one_hot(input_text=word, n=vocab_size) for word in Train3['selected_text']]
#One-hot encodes a text into a list of word indexes of size n.
X_Train2 = pad_sequences(sequences=one_hot_word1,
                              maxlen=len_sentence,
                              padding="pre")
#Pads sequences to the same length.
```

```python
X_Train2.shape
```

```python
#Test
```

```python
one_hot_word2 = [one_hot(input_text=word, n=vocab_size) for word in Test3['selected_text']]
#One-hot encodes a text into a list of word indexes of size n.
X_Test2 = pad_sequences(sequences=one_hot_word2,
                              maxlen=len_sentence,
                              padding="pre")
#Pads sequences to the same length.
```

```python
X_Test2.shape
```

```python
#Validation
```

```python
one_hot_word3 = [one_hot(input_text=word, n=vocab_size) for word in Val3['selected_text']]
#One-hot encodes a text into a list of word indexes of size n.
X_Val2 = pad_sequences(sequences=one_hot_word3,
                              maxlen=len_sentence,
                              padding="pre")
#Pads sequences to the same length.
```

```python
X_Val2.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, BatchNormalization, Conv2D, MaxPooling2D,Flatten ,Input,Embedding,LSTM
```

```python
#Assumption
embed_dim = 150

model = Sequential()
model.add(Embedding(input_dim=vocab_size, output_dim=embed_dim,input_length = X_Train2.shape[1]))
#vocabSize:input_dim (Integer. Size of the vocabulary, i.e. maximum integer index + 1.)
#embed_dim:output_dim (Integer. Dimension of the dense embedding.)
#input_length: Length of input sequences
model.add(Dropout(0.2))#explained below
#Dropout(): Let's define 2D input: [[1, 1, 1], [2, 2, 2]]. 
#Dropout will consider every element independently, and may result in something like [[1, 0, 1], [0, 2, 2]]
model.add(LSTM(200))
model.add(Dropout(0.3))
model.add(Dense(120,activation='relu'))
model.add(Dropout(0.3))        
model.add(Dense(6,activation='softmax'))
model.compile(loss = 'categorical_crossentropy', optimizer='adam',metrics = ['accuracy'])
print(model.summary())
```

```python
from keras.callbacks import EarlyStopping, ModelCheckpoint
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
Ohe = OneHotEncoder()
Y_Train2 = np.array(Train3['mood'])
Y_Train2 = Ohe.fit_transform(Y_Train2.reshape(-1,1)).toarray()

Y_Test2 = np.array(Test3['mood'])
Y_Test2 = Ohe.fit_transform(Y_Test2.reshape(-1,1)).toarray()

Y_Val2 = np.array(Val3['mood'])
Y_Val2 = Ohe.fit_transform(Y_Val2.reshape(-1,1)).toarray()
```

```python
batch_size = 30
M1=model.fit(X_Train2, Y_Train2, epochs = 5, batch_size=batch_size, verbose = 1,validation_data=(X_Val2,Y_Val2))
```

```python
#Model Train accuracy is 82% and Validation accuracy is 76%
#if we run for more epochs the train accuracy may increase but the validation accuray reamins constant
```

```python
score,acc = model.evaluate(X_Test2, Y_Test2, verbose = 1, batch_size = batch_size)
print("score: %.2f" % (score))
print("accuracy: %.2f" % (acc))
```

```python
def plot_performance1(model):
    plt.figure(figsize=(10, 5))
    plt.plot(model.history['accuracy'], 'b', label='train_accuracy')
    plt.plot(model.history['val_accuracy'], 'y', label='validation_accuracy')
    plt.legend()
    plt.title("Train-Validation accuracy vs Epochs")
    plt.grid(True)
```

```python
def plot_performance2(model):
    plt.figure(figsize=(10, 5))
    plt.plot(model.history['loss'], 'b--', label='train_loss')
    plt.plot(model.history['val_loss'], 'y--', label='validation_loss')
    plt.legend()
    plt.title("Train-Validation loss vs Epochs")
    plt.grid(True)
```

```python
plot_performance1(M1)
```

```python
plot_performance2(M1)
```
